Examine the categorical columns to determine if any value or column needs to be treated

```python
# Write a function to display the categorical columns with their unique values and check for redundant values

def display_categorical_info(df):
    categorical_cols = df.select_dtypes(include=['object']).columns
    print(f"Categorical columns: {list(categorical_cols)}\n")
    
    for col in categorical_cols:
        print(f"Column: {col}")
        unique_vals = df[col].unique()
        print(f"Unique values ({len(unique_vals)}): {unique_vals}")
        
        # Check for redundancies by comparing lowercase stripped versions
        normalized = df[col].astype(str).str.lower().str.strip()
        redundancies = normalized.value_counts()
        
        print("Normalized value counts for redundancy check:")
        print(redundancies, "\n")
```

7.2.3 Fit logistic regression model <font color="red">[3 Marks]</font>

```python
# Fit a logistic regression model on X_train after adding a constant and output the summary

# Reset indices for both y and X to ensure they are aligned
X_train_rfe_const_reset = X_train_rfe_const.reset_index(drop=True)
y_train_dummy_reset = y_train_dummy.reset_index(drop=True)
```

**Model Interpretation**

The output summary table  will provide the features used for building model along with coefficient of each of the feature and their p-value. The p-value in a logistic regression model is used to assess the statistical significance of each coefficient. Lesser the p-value, more significant the feature is in the model.

A positive coefficient will indicate that an increase in the value of feature would increase the odds of the event occurring. On the other hand, a negative coefficient means the opposite, i.e,  an increase in the value of feature would decrease the odds of the event occurring.



7.2.4 Evaluate VIF of features <font color="red">[3 Marks]</font>

```python
# Import 'variance_inflation_factor'
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Make a VIF DataFrame for all the variables present
vif_data = pd.DataFrame()
vif_data["feature"] = X_train_rfe_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_rfe_const.values, i)
                   for i in range(X_train_rfe_const.shape[1])]

# Display the VIF DataFrame
print(vif_data)
```

```python
# All VIFs are well below 5, most are even around 1.0, which is excellent. This logistic regression coefficients and p-values are trustworthy.No need to drop or transform variables due to multicollinearity.
```

7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# Define ROC function

def plot_roc_curve(y_true, y_prob):
  
    # Calculate ROC curve
    fpr, tpr, thresholds = metrics.roc_curve(y_true, y_prob)
    
    # Calculate AUC
    auc = metrics.roc_auc_score(y_true, y_prob)

    # Plot the ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {auc:.4f})')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal for random guessing
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

    return fpr, tpr, thresholds, auc
```

```python
y_prob.head()
```

```python
# Call the ROC function
y_true = results_df['Stayed_Actual']
y_prob = results_df['Stayed_Prob_Predicted']

# Call the function
fpr, tpr, thresholds, auc = plot_roc_curve(y_true, y_prob)

print(f"AUC Score: {auc:.4f}")
```
